Logical Operators :- Logical operators are used to combine conditional statements

```python
#  Operator  Description

#  and       Returns True if both statements are true
#  or        Returns True if one of the statements is true  
#  not       Reverse the result, returns False if the result is true
```

```python
# and operator :- Returns True if both statements are true.

x = 5
print( x > 3 and x < 10)
```

```python
# or operator :- Returns True if one of the statements is true.

x = 5
print( x < 3 or x < 10)
```

```python
# not operator :- Reverse the result, returns False if the result is true.

x = 5
print(not(x > 3 and x < 10))
```

Identity Operators :- Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

```python
#  Operator  Description

#  is        Returns True if both variables are the same object
#  is not    Returns True if both variables are not the same object
```

```python
# is operator :-  Returns True if both variables are the same object.

x = [1, 3, 5, 7, 9]
y = [1, 3, 5, 7, 9]
z = x

print(x is y)
print(x is z)
print(x == y)
```

```python
# is not operator :- Returns True if both variables are not the same object.

x = [1, 3, 5, 7, 9]
y = [1, 3, 5, 7, 9]
z = x

print(x is not y)
print(x is not z)
print(x != y)
```

Membership Operators :- Membership operators are used to test if a sequence is presented in an object.



```python
#  Operator  Description

#  in        Returns True if a sequence with the specified value is present in the object
#  not in    Returns True if a sequence with the specified value is not present in the object
```

```python
# in operator :- Returns True if a sequence with the specified value is present in the object.

x = [1, 3, 5, 7, 9]

print(5 in x)
```

```python
# not in operator :- Returns True if a sequence with the specified value is not present in the object.

x = [1, 3, 5, 7, 9]

print(5 not in x)
```

Bitwise Operators :- Bitwise operators are used to compare (binary) numbers.

```python
#  Operator   Name                  Description

#  &          AND                   Sets each bit to 1 if both bits are 1
#  |          OR                    Sets each bit to 1 if one of two bits is 1
#  ^          XOR                   Sets each bit to 1 if only one of two bits is 1
#  ~          NOT                   Inverts all the bits
#  <<         Zero fill left shift  Shift left by pushing zeros in from the right and let the leftmost bits fall off
#  >>         Signed right shift    Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off
```

```python
# AND Operator(&) :- Sets each bit to 1 if both bits are 1.

a = 60
b = 13
# 60 = 0011 1100
# 13 = 0000 1101
print(a & b)
# 12 = 0000 1100
```

```python
# OR Operator(|) :- Sets each bit to 1 if one of two bits is 1.

a = 60
b = 13
print(a | b)
# 61 = 0011 1101
```

```python
# XOR Operator(^) :- Sets each bit to 1 if only one of two bits is 1.

a = 60
b = 13
print(a ^ b)
# 49 = 0011 0001
```

```python
# NOT Operator(~) :- Invert all the bits.

a = 60
b = 13
print(~ a)
# -61 = 1100 0011
```

```python
# Zero fill left shift(<<) :- Shift left by pushing zeros in from the right and let the leftmost bits fall off.

a = 60
b = 13
print(a << 2)
# 240 = 1111 0000
```

```python
#  Signed right shift(>>) :- Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off.

a = 60
b = 13
print(a >> 2)
# 15 = 0000 1111
```
